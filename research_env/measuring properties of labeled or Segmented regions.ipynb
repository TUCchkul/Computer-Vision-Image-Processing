```python
from skimage import measure, io, img_as_ubyte
```

```python
import matplotlib.pyplot as plt
from skimage.color import label2rgb, rgb2gray
import numpy as np
```

```python
image=img_as_ubyte(rgb2gray(io.imread("images/cast_iron.jpg")))
image
```

```python
from skimage.filters import threshold_otsu
threshold=threshold_otsu(image)
```

```python
threshold
```

```python
label_img=measure.label(image<threshold, connectivity=image.ndim)
plt.imshow(label_img)
```

```python
props=measure.regionprops_table(label_img, image,
                               properties=['label','area','equivalent_diameter',
                                          'mean_intensity','solidity'])
```
